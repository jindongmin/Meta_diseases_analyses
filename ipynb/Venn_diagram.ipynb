```python
from matplotlib_venn import venn3, venn2
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#ASD all the differentially abundant microbes
table_dan = pd.read_table('../table/Dan_deseq2.txt')
set_Dan=set(table_dan['Unnamed: 0'])
table_wang = pd.read_table('../table/Wang_ASD_deseq2.txt')
set_wang = set(table_wang['Unnamed: 0'])
table_dan_n_wang = pd.read_table('../table/Dan_n_Wang_ASD_deseq2.txt')
set_dan_n_wang = set(table_dan_n_wang['Unnamed: 0'])
venn3([set_Dan, set_wang, set_dan_n_wang], ('set_Dan', 'set_wang', 'set_dan_n_wang'))
plt.show()
```

```python
#beneficial microbes in Healthy -- all ASD
#A positive log2 fold change for a comparison of A vs B means that abundance in A is larger in comparison to B
set_Dan = set(table_dan.loc[(table_dan['log2FoldChange'] >= 0)]['Unnamed: 0'])
set_Wang = set(table_wang.loc[(table_wang['log2FoldChange'] >= 0)]['Unnamed: 0'])
set_dan_n_wang = set(table_dan_n_wang.loc[(table_dan_n_wang['log2FoldChange'] >= 0)]['Unnamed: 0'])
venn3([set_Dan, set_Wang, set_dan_n_wang], ('set_Dan', 'set_wang', 'set_dan_n_wang'))
plt.show()
```

```python
#deleterious microbes in ASD
set_Dan = set(table_dan.loc[(table_dan['log2FoldChange'] <= 0)]['Unnamed: 0'])
set_Wang = set(table_wang.loc[(table_wang['log2FoldChange'] <= 0)]['Unnamed: 0'])
set_dan_n_wang = set(table_dan_n_wang.loc[(table_dan_n_wang['log2FoldChange'] <= 0)]['Unnamed: 0'])
venn3([set_Dan, set_Wang, set_dan_n_wang], ('set_Dan', 'set_wang', 'set_dan_n_wang'))
plt.show()
```

```python
set_Dan.intersection(set_Wang)
```
